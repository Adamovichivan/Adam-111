```python
import pandas as pd

# Створення даних для таблиці
data = {
    "Ім'я": ['Андрій', 'Олена', 'Михайло'],  # Використовуємо подвійні лапки для ключа
    'Вік': [30, 25, 40],
    'Місто': ['Київ', 'Львів', 'Одеса']
}

# Створення DataFrame з даних
df = pd.DataFrame(data)

# Збереження DataFrame у Excel-файл
df.to_excel('table.xlsx', index=False)

print("Excel файл успішно створено!")
```

```python
import pandas as pd

# Завантаження вихідного Excel-файлу
df = pd.read_excel('table.xlsx')

# Ваші фільтри для даних (наприклад, фільтрація за умовою)
filtered_df = df[df['Вік'] == 'some_value']  # Приклад фільтрації

# Збереження відфільтрованих даних у новий файл
filtered_df.to_excel('filtered_table.xlsx', index=False)

print("Відфільтровані дані успішно збережено у файл 'filtered_table.xlsx'")
```
